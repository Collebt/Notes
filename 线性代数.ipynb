## xy和幂函数的大小关系

## 考虑幂函数 $(x^a+y^a)^n$ 和 函数 $xy$之间的图像

---

> 我也不是很清楚$xy$ 和 $(x^a+y^a)^n$ 之间的关系，但按照题目我感觉是无论幂函数的次方系数如何变化，$xy$都存在路径比幂函数高阶无穷小

- 按照题目绘制 $g(x,y)=\frac{(x^2 + y^2 )^2}{xy}$的图像, 我们重点关注函数在接近0点的图像

```python
import numpy as np
import matplotlib.pyplot as plt
```

- 定义函数为$g(x,y)=\frac{(x^2 + y^2 )^2}{xy}$

```python
def fun(x,y):
    g = ((x**2 + y**2)**2 )/ (x*y)
    return g
```

```python
#set input meshgrid
X,Y = np.meshgrid(np.linspace(-2,2,30), np.linspace(-2,2,30))#定义自变量x,y的范围为(-10,10),等间隔取30个点
#calculate the g(x,y) output
G = fun(X,Y) #这里我们用上上面定义好的函数计算输出
```

 - 得到输出后进行绘图，这里用比较通用的绘图包matplotlib

```python
#开始画花里胡哨的图
fig = plt.figure() #创建绘图对象fig
ax = plt.axes(projection='3d') #建立三维坐标系图
ax.plot_surface(X,Y,G,cmap='viridis',edgecolor='none')
ax.contour(X,Y,G,zdir='z', offset=-200, cmap='viridis')#在三维曲面下面画等高线图
ax.set_zlim(-200,100)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('f(x,y)')
plt.show() #把图打在屏幕上
```

> 好像从上图也没有看出来在$y=x^4$那个路径趋向于0是上升到正无穷的,
但在下面的等高线中可以看到一堆$y=\pm x^n, （n>4或n<\frac{1}{4}）$的路径,可以理解为这些路径上的数值都异常的高，在接近0的地方开始拉成一条直线（十字形）是因为路径很窄导致的。

但实际上在任何幂函数$f(x,y) = \frac{(x^n +y^m)^k}{x^ay^b}$中,总能选择出路径$y=x^\frac{nk-a+1}{b}$ 或者 $x=y^\frac{mk-b+1}{a}$使趋近于0达到正无穷$\infty$
